**1. Explain the key features of Python that make it a popular choice for programming.**

 - Python is one of the most popular programming languages because of the following features:

 **Easy to Learn and Use:** Python has simple and readable syntax similar to the English language, which makes it easy for beginners.

**Interpreted Language:** Python executes code line by line, which makes debugging easier.

**High-Level Language:** Python handles memory management automatically, so the programmer does not need to worry about low-level details.

**Dynamically Typed:** Variable types are decided at runtime, so no need to declare data types explicitly.

**Large Standard Library:** Python provides many built-in libraries for file handling, math operations, data analysis, web development, etc.

**Platform Independent:** Python code can run on Windows, Linux, and macOS without modification.

**Supports Multiple Programming Paradigms:** Python supports procedural, object-oriented, and functional programming.

**3. Compare and contrast mutable and immutable objects in Python with examples**.

- Mutable Objects:
* Objects whose values can be changed after creation.
*  Changes affect the original object.

**Examples:** List, Dictionary, Set.


```python
my_list = [1, 2, 3]
my_list[0] = 10
print(my_list)
```

 - Immutable Objects:


*  Objects whose values cannot be changed after creation.
*  A new object is created if modification is attempted.

**Examples:** Integer, Float, String, Tuple.

```python
my_string = "Hello"
# my_string[0] = "h"  # This will give an error
print(my_string)
```

**4. Discuss the different types of operators in Python and provide examples of how they are used.**

**5. Explain the concept of type casting in Python with examples**.
  
  - Type casting means converting one data type into another.

  **Types of Type Casting:**
  
   **1. Implicit Type Casting:** Python automatically converts data type.

```python
x = 10
y = 2.5
z = x + y
print(z)
```

**2. Explicit Type Casting:** Programmer manually converts data type.

```python
a = "100"
b = int(a)
print(b + 10)
```

**7. Describe the different types of loops in Python and their use cases with examples.**
 - Loops are used to execute a block of code repeatedly.
   
    **1. For Loop:** Used when the number of iterations is known.

```python
for i in range(1, 6):
    print(i)
```

 **2. While Loop:** Used when the number of iterations is not known.

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

 **3. Nested Loop:** Loop inside another loop.

```python
for i in range(1, 3):
    for j in range(1, 3):
        print(i, j)
```
